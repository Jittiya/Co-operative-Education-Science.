## แทนด้วยค่ากลางของคอลัมน์นั้น (Mean)

ในกรณีนี้จะหาค่ากลาง(mean) ของแต่ละคอลัมน์นั้น แล้วนำไปเติมในช่องที่เป็นค่าว่าง

```python
dataFore[['จำนวนผู้ว่างงาน (พันคน)']].mean()
```

```python
# หา mean
avg_unemploy = dataFore[['จำนวนผู้ว่างงาน (พันคน)']].mean()
# เติม mean ลงไปแทนค่า missing
dataFore[['จำนวนผู้ว่างงาน (พันคน)']] = dataFore[['จำนวนผู้ว่างงาน (พันคน)']].fillna(avg_unemploy)
```

```python
dataFore[['ร้อยละ']].mean()
```

```python
# หา mean
avg_Per = dataFore[['ร้อยละ']].mean()
# เติม mean ลงไปแทนค่า missing
dataFore[['ร้อยละ']] = dataFore[['ร้อยละ']].fillna(avg_Per)
```
