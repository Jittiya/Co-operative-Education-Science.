# 4 Techniques to Handle Missing values in Time Series Data

*True* for missing values, \
*False* for not missing values:

## 1. Last Observation Carried Forward (LOCF)

```python
df_LOCF = data1
df_LOCF
```

```python
df_LOCF['จำนวนผู้ว่างงาน (พันคน)'] = df_LOCF['จำนวนผู้ว่างงาน (พันคน)'].ffill()
```

```python
df_LOCF['ร้อยละ'] = df_LOCF['ร้อยละ'].ffill()
```

```python
df_LOCF
```

```python
df_LOCF.isnull().any()
```

## 2. Next Observation Carried Backward (NOCB)

```python
df_NOCB = data2
df_NOCB
```

```python
df_NOCB['จำนวนผู้ว่างงาน (พันคน)'] = df_NOCB['จำนวนผู้ว่างงาน (พันคน)'].bfill()
```

```python
df_NOCB['ร้อยละ'] = df_NOCB['ร้อยละ'].bfill()
```

```python
df_NOCB
```

## 3. Rolling Statistics

```python
df_ROLL = data3
df_ROLL
```

### **Simple Moving Average**

```python
df_ROLL_SMA = df_ROLL
df_ROLL_SMA
```

```python
df_ROLL_SMA['SMA Total'] = df_ROLL_SMA['จำนวนผู้ว่างงาน (พันคน)'].rolling(window=5).mean()
```

```python
df_ROLL_SMA['SMA Per'] = df_ROLL_SMA['ร้อยละ'].rolling(window=5).mean()
```
